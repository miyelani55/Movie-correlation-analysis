Importing dependencies

```python
import pandas as pd
import numpy as np
```

Data preprocessing

```python
my_path = "/content/sample_data/spam_dataset.csv"
df = pd.read_csv(my_path)
df.head(10)
```

Check the size of our data

```python
print(f"Number of records is: {df.shape[0]}")
print(f"Number of field is: {df.shape[1]}")
```

Getting a picture of how out data is like, nulls, datatype,features

```python
df.info()
```

Check the data after renaming feature

```python
df=df.rename(columns={'is_spam':'label'})
df.head()
```

Assigning variables

```python
x = df['message_content']
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
```

Splitting the data

```python
X_train, X_test, Y_train, Y_test =train_test_split(x,y, test_size=0.2, random_state=42)
```

Check the size of train and test data

```python
print (f"Size of train data is: {X_train.shape}")
print (f"Size of test data is: {X_test.shape}")
```

Transform text data to feature vector that can be used as input to the model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
feature_extraction=TfidfVectorizer(min_df=1, stop_words='english',lowercase=True)
X_train_tfidf =feature_extraction.fit_transform(X_train)
X_test_tfidf = feature_extraction.transform(X_test)
```

import logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

Train model

```python
model = LogisticRegression()
model.fit(X_train_tfidf, Y_train)
```

import evaluation metrics

```python
from sklearn.metrics import accuracy_score, f1_score
```

Prediction on train data

```python
Y_pred = model.predict(X_train_tfidf)

accuracy_on_train_data = accuracy_score(Y_pred, Y_train)
f1_score_trained_data = f1_score(Y_pred, Y_train)
```

```python
print (accuracy_on_train_data)
print (f1_score_trained_data)
```

Prediction on test data

```python
Y_pred = model.predict(X_test_tfidf)

accuracy_on_test_data = accuracy_score(Y_pred, Y_test)
f1_score_test_data = f1_score(Y_pred, Y_test)
print (accuracy_on_test_data)
print (f1_score_test_data)
```

Building predictive system

```python
input_data=["Hello, how are you doing today?",    "Meeting tomorrow at 3pm in conference room"]
input_data_tfidf = feature_extraction.transform(input_data)
```

```python

prediction = model.predict (input_data_tfidf)

if prediction[0]==1:
  print("Message is a spam")

else:
  print("Not a spam")
```
